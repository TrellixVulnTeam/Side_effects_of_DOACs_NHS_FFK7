# Binary Events

```python
def get_events_on_dose(dose_df):
    kk = dose_df["Bleeding_Risk"].value_counts()
    #print(kk)
    jj = dose_df["VTE_Risk"].value_counts()
    #print(jj)
    
    frame = {'Bleeding_Riks': pd.Series(kk), 'VTE_Risk': pd.Series(jj)}
    events_df = pd.DataFrame(frame)
    
    events_df['Deceased_in_EPR'] = pd.Series(dose_df["Deceased_in_EPR"].value_counts())
    events_df['Has_Bleeding_Event_(or_Stroke)_on_Spell'] = pd.Series(dose_df["Has_Bleeding_Event_(or_Stroke)_on_Spell"].value_counts())
    events_df['Has_Stroke_on_Spell'] = pd.Series(dose_df["Has_Stroke_on_Spell"].value_counts())
    events_df['Has_Thrombosis_on_Spell'] = pd.Series(dose_df["Has_Thrombosis_on_Spell"].value_counts())
    
    
    print(events_df.shape)
    print(events_df.head())
    return events_df
    
```

```python
Events_apixaban_5_df = get_events_on_dose(apixaban_5_df)
Events_apixaban_10_df = get_events_on_dose(apixaban_10_df)
Events_apixaban_205_df = get_events_on_dose(apixaban_205_df)
Events_dabigatran_150_df = get_events_on_dose(dabigatran_150_df)
Events_dabigatran_110_df = get_events_on_dose(dabigatran_110_df)
Events_dabigatran_75_df = get_events_on_dose(dabigatran_75_df)
Events_edoxaban_30_df = get_events_on_dose(edoxaban_30_df)
Events_edoxaban_60_df = get_events_on_dose(edoxaban_60_df)
Events_rivaroxaban_20_df = get_events_on_dose(rivaroxaban_20_df)
Events_rivaroxaban_15_df = get_events_on_dose(rivaroxaban_15_df)
Events_rivaroxaban_10_df = get_events_on_dose(rivaroxaban_10_df)
```
